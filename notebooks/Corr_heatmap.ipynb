```python
### heat map
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_WHO = pd.read_csv("../data/processed/WHO_data_cleaned.csv")
# df_WHO.head()
```

```python
sns.heatmap(df_WHO[["Density Med Fac (100,000 pop)", "Medical Doctors per 10k inhabitants", "Infant mortality per 1000 live births"]].corr(), 
            annot = True, cmap = 'PiYG')
plt.xticks(rotation=45)
```
