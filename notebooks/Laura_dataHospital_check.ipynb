```python
import pandas as pd
```

```python
df = pd.read_csv('../data/external/data_hospitals_100k_who.csv')
```

```python
df.head()
```

```python
df_cleaned = df[['ParentLocation','Location','Period','Value']]
df_cleaned = df_cleaned.rename(columns = {'Location':'Country', 'Value':'Ratio_hosp_100k'})
df_cleaned
```

```python
df_cleaned[df_cleaned['ParentLocation'] == 'Africa']
```

```python
countries = [
    "Angola",
    "Burundi",
    "Cameroon",
    "Central African Republic",
    "Chad",
    "Democratic Republic of the Congo",
    "Equatorial Guinea",
    "Gabon",
    "Republic of the Congo",
    "Rwanda",
    "São Tomé and Príncipe"
]
```

```python
# Filtering for central Africa
df_central = df_cleaned[df_cleaned['Country'].isin(countries)]
df_central = df_central.iloc[:, 1:]
df_central
```

```python
# comparison with data from Wikipedia
df_mfp = pd.read_csv('medical_facilities_population.csv')
df_mfp
```

```python
df_merged = pd.merge(df_mfp,df_central, how = 'outer')
```

```python
df_merged
```

```python
df_merged['hospitals'] = df_merged['Population (1,000)']/100*df_merged['Ratio_hosp_100k']
```

```python
df_merged
```

```python
#Extracting number hospitals from Wikipedia data

df_list_hospital = pd.read_csv('hospitals_list.csv')
df_list_hospital
```

```python
df_grouped = df_list_hospital.groupby('Country').size().reset_index()
df_grouped
```

```python
df_grouped = df_grouped.rename(columns = {0:'n_hospitals_wiki'})
df_grouped
```

```python
df_merged = pd.merge(df_merged,df_grouped, how='outer')
df_merged
```

```python
#Estracting data from external website csv
df_external = pd.read_csv('../data/external/who-cds-gmp-2019-01-eng.csv')
df_external
```

```python
df_external_central = df_external[df_external['Country'].isin(countries)]
df_external_central
```

```python
df_external_central.to_csv('../data/processed/central_Africa_hospitals.csv')
```

```python
df_extracted= df_external_central.groupby('Country').size().reset_index().rename(columns = {0:'n_hosp_2019'})
df_extracted
```

```python
df_merged = pd.merge(df_merged,df_extracted, how='outer')
df_merged
```
