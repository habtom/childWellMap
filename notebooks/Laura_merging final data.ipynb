```python
import pandas as pd
```

```python
# Uploading file WHO_data_central.csv
df = pd.read_csv('../data/WHO_data_central.csv')
```

```python
df
```

```python
df=df.rename(columns = {'Location':'Country'})
df=df.iloc[:,3:]
df
```

```python
df = df.drop(columns = ['Period_doctor_data', 'Period'])
df
```

```python
#Uploading file from Wiki
df_2 = pd.read_csv('medicalFacilitiesPopulation2023.csv')
df_2['Country'] = df_2['Country'].replace({'Republic of the Congo':'Congo', 'São Tomé and Príncipe' : 'Sao Tome and Principe'})
df_2
```

```python
df_cleaned = pd.merge(df_2,df, how= 'outer')
df_cleaned
```

```python
df_cleaned.to_csv('../data/processed/WHO_data_cleaned.csv', index=False, encoding='utf-8')
```
